```python
numb_of_categories = len(df['tshirt_category'].unique())
print(f'{numb_of_categories}')
```

```python
df['tshirt_category'] = df['tshirt_category'].astype('string')
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: x.replace('Wh ', 'White ').replace('Bl ', 'Black ').replace('Tshirt ', 'T-Shirt ').replace('Hoodie','Different Hoodie Different').replace(' Shirt', ' Shirt Different'))
print(df['tshirt_category'].unique())
```

```python
df['tshirt_type'] = df['tshirt_category'].apply(lambda x: x.split(' ')[1])
df['tshirt_sex'] = df['tshirt_category'].apply(lambda x: x.split(' ')[2])
df['tshirt_colour'] = df['tshirt_category'].apply(lambda x: x.split(' ')[0])
df
```

```python
df_date = df[(df['order_date'] > datetime.date(2014,12,31)) & (df['order_date'] < datetime.date(2016,8,2))].sort_values('order_date')
df_date.head(10)
```
