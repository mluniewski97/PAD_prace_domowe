```python
import pandas as pd
```

```python
#Zad 1
df=pd.read_csv('/Users/michalluniewski/Documents/PAD/03_PAD/Zadania_domowe/PAD_03_PD.csv', sep=';')
df
```

```python
#Replace missing values with the most frequent value in the column
cols = ['Country','owns_car', "owns_TV", 'owns_house', 'owns_Phone', 'Age', 'gender']
df[cols]=df[cols].fillna(df.mode().iloc[0])
df

```

```python
df.groupby('Country').count()
```

```python
#Zad 2
df['Owned_Goods'] = df.owns_car + df.owns_TV + df.owns_house + df.owns_Phone
df
```

```python
#Zad 3
df.groupby('gender').mean().Owned_Goods.round(2)
```

```python
#Zad 4
CountryMean = df.groupby('Country').mean().Owned_Goods.round(2)
AgeMin = df.groupby('Country').min().Age
NewDf = pd.merge(CountryMean, AgeMin, how='outer', on='Country')
NewDf
```
