Zad 4

```python
df4 = pd.read_csv('/Users/michalluniewski/Documents/PAD/04_PAD/Zadanie domowe/Zadanie_4.csv', sep=';', usecols=[0,1,2,3,])
df4.head(5)
```

```python
df4["Date"] = pd.to_datetime(df4['DateTime']).dt.date
df4["Time"] = pd.to_datetime(df4["DateTime"]).dt.time
df4.head(5)
```

```python
df4_min = df4.groupby(['DoctorID','Date','City','Type']).agg({'Time':'min'})
df4_max = df4.groupby(['DoctorID','Date','City','Type']).agg({'Time':'max'})
df4_min.rename(columns={'Time':'The doctor takes from'},inplace=True)
df4_max.rename(columns={'Time':'The doctor admits to'},inplace=True)
df4_final = pd.concat([df4_min,df4_max], join='inner', axis = 1)
df4_final
```

```python
df4_final.to_csv('out.csv')
```
